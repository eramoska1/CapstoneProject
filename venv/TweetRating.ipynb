```python
# Evelina Ramoskaite
import pandas as pd
import csv
# Tweet Rating Code
fileName= input('FileName:')
stock= input('hashtag ticker for file:')
data = pd.read_csv(fileName)
print('There are {} records in this dataset.'.format(len(data)))
print('to quit, type quit.')
# Decoding the text columns
import ast
data.tweet_text = data['tweet_text'].apply(ast.literal_eval).str.decode("utf-8")
data[' username']= data[' username'].apply(ast.literal_eval).str.decode("utf-8")

ratings = []
count= 0
for i in data.tweet_text:
    print(i)
    rating = input('Rate(1/2/3):')
    ratings.append(rating)
    count += 1
    if count//50:
        print('count: {} , total left: {}'.format(count,len(data)-count))
    if rating == 'quit':
        dict = {'rating': ratings}
        ratingData = pd.DataFrame(dict)
        ratingData.to_csv('rated_{}_inc.csv'.format(stock))
        break
data['target'] =ratings
data.target
data.to_csv('rated_{}.csv'.format(stock))
```
